```python
data = ('Linear Double Seasonal', 'Non-Linear Double Seasonal', 'Kaggle Electric Production', 'Kaggle Wines Sales')
metrics = {
    'R2 Score GP': (0.933, 0.955, 0.843, 0.964),
    'R2 Score SARIMAX': (0.957, 0.951,  0.856, 0.968),
}

x = np.arange(len(data))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0
fig, ax = plt.subplots(layout='constrained')
fig.set_size_inches(10, 6, forward=True)

for attribute, measurement in metrics.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('R2 Score')
ax.set_title('Data')
ax.set_xticks(x + width, data)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 1.2)

plt.show()
```
